```python
#split the dataset to train valid sets
import os, shutil
from sklearn.model_selection import train_test_split

file = []
for i in os.listdir("/home/arvind/Python/Fooball-CV-analysis/training/SOD_Dataset/images"):
    file.append(i.split(".")[0])

train, valid = train_test_split(file, train_size = 0.8, test_size = 0.2, random_state = 42)
os.makedirs("dataset/train/images", exist_ok=True)
os.makedirs("dataset/train/labels", exist_ok=True)
os.makedirs("dataset/valid/images", exist_ok=True)
os.makedirs("dataset/valid/labels", exist_ok=True)

for i,j in zip(os.listdir("/home/arvind/Python/Fooball-CV-analysis/training/SOD_Dataset/images"), os.listdir("/home/arvind/Python/Fooball-CV-analysis/training/SOD_Dataset/annotations")):
    if i.split(".")[0] in train:
        shutil.copy(os.path.join("/home/arvind/Python/Fooball-CV-analysis/training/SOD_Dataset/images", i),"dataset/train/images")
    if j.split(".")[0] in train:
        shutil.copy(os.path.join("/home/arvind/Python/Fooball-CV-analysis/training/SOD_Dataset/annotations", j),"dataset/train/labels")
    if i.split(".")[0] in valid:
        shutil.copy(os.path.join("/home/arvind/Python/Fooball-CV-analysis/training/SOD_Dataset/images", i),"dataset/valid/images")
    if j.split(".")[0] in valid:
        shutil.copy(os.path.join("/home/arvind/Python/Fooball-CV-analysis/training/SOD_Dataset/annotations", j),"dataset/valid/labels")
```
